```python
from bosonicplus.states.from_sf import prepare_gkp_bosonic
from bosonicplus.plotting import plot_wigner_marginals
from bosonicplus.states.nongauss import prepare_gkp_coherent
import numpy as np
from matplotlib import pyplot as plt
from thewalrus.symplectic import xxpp_to_xpxp, rotation
```

```python
#GKP 0
state = prepare_gkp_bosonic([0,0],0.05)
x = np.linspace(-10,10,200)

W = state.get_wigner(x,x)
plot_wigner_marginals(W.real, x,x, r'$|0\rangle_{\text{GKP}}$')
plt.savefig('GKP0.png',dpi=800)
plt.show()
```

```python
#GKP +
state = prepare_gkp_bosonic([0,0],0.05)
state.apply_symplectic(xxpp_to_xpxp(rotation(np.pi/2)))
x = np.linspace(-10,10,200)

W = state.get_wigner(x,x)
plot_wigner_marginals(W.real, x,x, r'$|+\rangle_{\text{GKP}}$')
plt.savefig('GKP+.png',dpi=800)
plt.show()
```

```python
#GKP 1
state = prepare_gkp_bosonic([np.pi,0],0.05)
x = np.linspace(-10,10,200)

W = state.get_wigner(x,x)
plot_wigner_marginals(W.real, x,x, r'$|1\rangle_{\text{GKP}}$')
plt.savefig('GKP1.png',dpi=800)
plt.show()
```
